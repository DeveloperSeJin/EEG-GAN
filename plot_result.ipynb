```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 준비
metrics = ["BLEU-1", "BLEU-2", "BLEU-3", "BLEU-4", "ROUGE-1 F", "ROUGE-2 F", "ROUGE-L F", "WER", "CER"]
baseline = [0.15, 0.03, 0.01, 0.00, 0.11, 0.01, 0.09, 1.15, 0.86]
eeg_gan = [0.16, 0.03, 0.01, 0.00, 0.12, 0.01, 0.10, 1.19, 0.95]
baseline_tf = [0.37, 0.19, 0.09, 0.05, 0.25, 0.04, 0.23, 0.81, 0.63]
eeg_gan_tf = [0.40, 0.22, 0.13, 0.08, 0.29, 0.08, 0.27, 0.78, 0.60]

# Plot 생성
fig, axes = plt.subplots(1, 2, figsize=(14, 6))  # 두 개의 subplot 생성

# Baseline vs EEG-GAN
x = np.arange(len(metrics))
width = 0.35
axes[0].bar(x - width/2, baseline, width, label='Baseline')
axes[0].bar(x + width/2, eeg_gan, width, label='EEG-GAN')
axes[0].set_title('Baseline vs EEG-GAN')
axes[0].set_xticks(x)
axes[0].set_xticklabels(metrics)
axes[0].legend()
axes[0].set_ylabel('Scores')
axes[0].set_xlabel('Metrics')

# Baseline with TF vs EEG-GAN with TF
axes[1].bar(x - width/2, baseline_tf, width, label='Baseline + TF')
axes[1].bar(x + width/2, eeg_gan_tf, width, label='EEG-GAN + TF')
axes[1].set_title('Baseline with TF vs EEG-GAN with TF')
axes[1].set_xticks(x)
axes[1].set_xticklabels(metrics)
axes[1].legend()
axes[1].set_xlabel('Metrics')

# 그래프 레이아웃 설정 및 출력
plt.tight_layout()
plt.show()
```
